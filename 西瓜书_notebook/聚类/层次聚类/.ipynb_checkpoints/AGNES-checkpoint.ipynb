### 层次聚类



### Code --- AGNES

```python
# 导包

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 对中文的支持
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 加载数据 西瓜数据集4.0

df = pd.read_csv('data4.0.csv')
df
```

```python
# AGNES 算法

def dist(x,y):
    # 欧式距离
    d = np.sqrt(np.sum((x-y)**2,axis=1))
    
    return d
    
def AGNES(D,k):
    """
    D:样本数据集 
    k:聚类簇数
    return:返回划分簇 C
    """
    cp = np.copy(D)
    m = len(cp)
    
    C = {} # 簇划分集合
    # 初始聚类簇
    for j in range(m):
        C[str(j)] = {j}
    
    M = np.ones((m,m))*np.inf
    
    def dmax(x1,x2):
        m1 = len(x1)
        m2 = len(x2)
        
        d = [ np.max(dist(np.tile(x1[i,:],(m2,1)),x2)) for i in range(m1)]
    
        return max(d)
        
    # 初始化簇距离矩阵
    for i in range(m):
        for j in range(m):
            if i!=j:
                ci = cp[list(C[str(i)]),:]
                cj = cp[list(C[str(j)]),:]
                M[i][j] = dmax(ci,cj)
                M[j][i] = M[i][j]
    
    # 设置当前聚类簇个数 q = m
    q = m
    
    while q>k:
        # 找出距离最近的两个类簇 Ci* 和 Cj* , i*<j*

        arg1,arg2 = np.where(M==M.min())
        arg_i,arg_j = sorted([arg1[0],arg2[0]])

        # 合并Ci 和 Cj
        C[str(arg_i)] = C[str(arg_i)] | C[str(arg_j)]

        for j in range(arg_j+1,q):
            # 将聚类簇 Cj 重新编号为 C(j-1)
            C[str(j-1)] = C[str(j)]
        
        # 将最后一个簇划分删除
        C.pop(str(q-1))
        
        # 删除距离矩阵 M 的第 j* 行与第 j* 列
        M = np.delete(M,arg_j,axis=0)
        M = np.delete(M,arg_j,axis=1)
        for j in range(q-1):
            if j!=arg_i:
                ci = cp[list(C[str(arg_i)]),:]
                cj = cp[list(C[str(j)]),:]
                M[arg_i,j] = dmax(ci,cj)
                M[j][arg_i] = M[arg_i][j]
        q = q-1
    
    return C
```

```python
# show figure
def show(D,C,title):
    
    cp = np.copy(D)
    c = ['c', 'b', 'g', 'r', 'm', 'y', 'k', 'w']
    plt.figure()
    for i in range(len(C)):
        x1 = cp[list(C[str(i)]),0]
        x2 = cp[list(C[str(i)]),1]
        plt.scatter(x1, x2, alpha=0.4, c=c[i])
        plt.xlabel('密度')
        plt.ylabel('含糖率')
        plt.title(title)
    plt.legend(['C'+str(i) for i in range(len(C))])
```

```python
dt = df.iloc[:,1:].to_numpy()
```

```python
C = AGNES(D=dt,k=7)
C
```

```python
Kc = [7,6,5,4]

for k in Kc:
    C =AGNES(D=dt,k=k)
    title = '聚类簇数 k='+str(k)
    show(D=dt,C=C,title=title)
```
