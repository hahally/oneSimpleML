### Code --- AGNES

```python
# 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#  4.0

df = pd.read_csv('data4.0.csv')
df
```

```python
# AGNES 

def dist(x,y):
    # 
    d = np.sqrt(np.sum((x-y)**2,axis=1))
    
    return d
    
def AGNES(D,k):
    """
    D: 
    k:
    return: C
    """
    cp = np.copy(D)
    m = len(cp)
    
    C = {} # 
    # 
    for j in range(m):
        C[str(j)] = {j}
    
    M = np.ones((m,m))*np.inf
    
    def dmax(x1,x2):
        m1 = len(x1)
        m2 = len(x2)
        
        d = [ np.max(dist(np.tile(x1[i,:],(m2,1)),x2)) for i in range(m1)]
    
        return max(d)
        
    # 
    for i in range(m):
        for j in range(m):
            if i!=j:
                ci = cp[list(C[str(i)]),:]
                cj = cp[list(C[str(j)]),:]
                M[i][j] = dmax(ci,cj)
                M[j][i] = M[i][j]
    
    #  q = m
    q = m
    
    while q>k:
        #  Ci*  Cj* , i*<j*

        arg1,arg2 = np.where(M==M.min())
        arg_i,arg_j = sorted([arg1[0],arg2[0]])

        # Ci  Cj
        C[str(arg_i)] = C[str(arg_i)] | C[str(arg_j)]

        for j in range(arg_j+1,q):
            #  Cj  C(j-1)
            C[str(j-1)] = C[str(j)]
        
        # 
        C.pop(str(q-1))
        
        #  M  j*  j* 
        M = np.delete(M,arg_j,axis=0)
        M = np.delete(M,arg_j,axis=1)
        for j in range(q-1):
            if j!=arg_i:
                ci = cp[list(C[str(arg_i)]),:]
                cj = cp[list(C[str(j)]),:]
                M[arg_i,j] = dmax(ci,cj)
                M[j][arg_i] = M[arg_i][j]
        q = q-1
    
    return C
```

```python
# show figure
def show(D,C,title):
    
    cp = np.copy(D)
    c = ['c', 'b', 'g', 'r', 'm', 'y', 'k', 'w']
    plt.figure()
    for i in range(len(C)):
        x1 = cp[list(C[str(i)]),0]
        x2 = cp[list(C[str(i)]),1]
        plt.scatter(x1, x2, alpha=0.4, c=c[i])
        plt.xlabel('')
        plt.ylabel('')
        plt.title(title)
    plt.legend(['C'+str(i) for i in range(len(C))])
```

```python
dt = df.iloc[:,1:].to_numpy()
```

```python
C = AGNES(D=dt,k=7)
C
```

```python
Kc = [7,6,5,4]

for k in Kc:
    C =AGNES(D=dt,k=k)
    title = ' k='+str(k)
    show(D=dt,C=C,title=title)
```
