### k 

 $D={x_1,x_2,...,x_m}$,$k-means$  $C={C_1,C_2,...C_k}$ 
$$
E = \sum_{i=1}^{k}\sum_{x\in C_i}||x-\mu_i||_2^2\\
\mu_i = \frac{1}{|C_i|}\sum_{x\in C_i}x
$$

$\mu_i$  $C_i$ $E$ 



### Code --- K-means

```python
# 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
dt = df.iloc[:,1:].to_numpy(copy=True)
```

```python
plt.scatter(dt[:,0],dt[:,1],c='g',alpha=0.5)
plt.xlabel('')
plt.ylabel("")
plt.show()
```

```python
# k-means 


def K_means(D, k, expoch):
    """
    D: np.array
    k:
    expoch:
    return: C = C{C1,C2,..,Ck}
    """
    cp_D = np.copy(D)
    m = cp_D.shape[0]

    #  D  k  {u1,u2,..,uk}
    miu = cp_D[[5, 11, 23], :]
#     import random
#     sli = random.sample(range(0,30),k)
#     miu = cp_D[sli, :]
    C = {}
    for step in range(expoch):
        #  C
        for n in range(k):

            C[str(n)] = set()

        for j in range(m):
            #  xj  miu dist_ji
            xj = np.tile(cp_D[j], (k, 1))
            dist = np.sqrt(np.sum((xj - miu)**2, axis=1))
            #  xj 
            min_dist = np.min(dist)
            argmin = np.argmin(dist)
            C[str(argmin)].add(j)
        
        
        count = 0
        c = ['b','g','r']
        plt.figure()
        for i in range(k):
            # 
            new_miu_i = np.sum(cp_D[list(C[str(i)]),:], axis=0) / len(C[str(i)])
            plt.scatter(new_miu_i[0],new_miu_i[1],s=80,c=c[i],marker='x')
            plt.scatter(cp_D[list(C[str(i)]),0],cp_D[list(C[str(i)]),1],c=c[i],alpha=0.5)
            plt.title("%d"%(step+1))
            plt.xlabel('')
            plt.ylabel('')
            
            flag = np.sum((np.abs(miu[i] - new_miu_i) <= 1e-10) + 0)
            if flag < len(new_miu_i):
                miu[i] = new_miu_i
            else:
                count = count + 1
        if count == k:
            break
        print("step %d :"%step)
        print('miu:',miu)
        print('C',C)
        print('='*30)
    return C,miu
```

```python
C,miu = K_means(D=dt,k=3,expoch=10)
```
