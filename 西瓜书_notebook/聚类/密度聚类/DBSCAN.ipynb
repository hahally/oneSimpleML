### 密度聚类

### Code --- DBSCAN



```python
# 导包

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 对中文的支持
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 加载数据 西瓜数据集4.0

df = pd.read_csv('data4.0.csv')
df
```

```python
# DBSCAN 算法
import random

def DBSCAN(D,epsilon,minpts):
    """
    D: 样本集 {x1,x2,...,xm}
    epsilon,minpts: 邻域参数
    return：返回划分簇 C 样本下标组成的集合
    """
    cp = np.copy(D)
    m = len(cp)
    C = {}
    c = set()
    
    # 初始化核心对象集合：Ω = 空
    Omega = set()
    
    for j in range(m):
        # 确定样本 xj 的 ε-邻域 Nε(xj)
        xj = np.tile(cp[j,:],(m,1))
        dist = np.sqrt(np.sum((xj-cp)**2,axis=1))
        N_args = np.where(dist<=epsilon)

        if len(N_args[0])>=minpts:
            # 将样本 xj 加入核心集合：Ω
            Omega.add(j)
    # 初始化聚类簇数 k = 0
    k = 0
    
    # 初始化未访问的样本集合 F = D
    F = set(range(m))
    while Omega:
        # 记录当前未访问样本集合
        F_old = F.copy()
        
        # 随机选取一个核心对象 o in Ω，初始化队列 Q=<o>
        arg = random.sample(range(0,len(Omega)),1)
#         arg = [3]
        o = list(Omega)[arg[0]]
        Q = []
        Q.append(o)
        F.remove(o)
        while Q:
            # 取出队列 Q 中的首个样本 q
            q = Q.pop(0)
            dist = np.sqrt(np.sum((cp[q]-cp)**2,axis=1))
            N_args = np.where(dist<=epsilon)
            if len(N_args[0])>=minpts:
            
                # 令 delta = Nε(q) & F
                delta = set(N_args[0]) & F
                
                # 将 delta 中的样本加入队列 Q
                Q = Q+list(delta)
                
                for i in delta:
                    F.remove(i)
                    
        # 生成聚类簇 Ck = F_old\F
        k = k+1
        C[str(k)] = F_old.symmetric_difference(F_old&F)
        c = c|C[str(k)]
        Omega.symmetric_difference_update(Omega&C[str(k)])
    
    # 噪声样本
    all_dt = set(range(m))
    noise = all_dt.symmetric_difference(c)
    
    noise
    return C,noise
```

```python
# show figure
def show(D,C,title):
    
    cp = np.copy(D)
    c = ['c', 'b', 'g', 'r', 'm', 'y', 'k', 'w']
    plt.figure()
    mk = 'o'
    for i in range(1,len(C)+1):
        x1 = cp[list(C[str(i)]),0]
        x2 = cp[list(C[str(i)]),1]
        if i==len(C):
            mk = 'x'
        plt.scatter(x1, x2, alpha=0.4, c=c[i],marker=mk)
        plt.xlabel('密度')
        plt.ylabel('含糖率')
        plt.title(title)
    plt.legend(['C'+str(i+1) for i in range(len(C)-1)]+['noise'])
```

```python
dt = df.iloc[:,1:].to_numpy()
C,noise = DBSCAN(dt,epsilon=0.11,minpts=5)
C,noise
```

```python
C['5'] = noise
show(dt,C,'DBSCAN')
```
