### k-邻近学习

k 邻近($k-Nearest$ $Neighbor$ 简称 $kNN$) 学习是一种常用的监督学习方法(惰性学习)。

>学习机制：给定测试样本，基于某种距离度量找出训练集中与其最靠近的 $k$ 个训练样本，然后基于这$k$ 个“邻居”的信息来进行预测。通常，在分类任务中可使用“投票法”，即选择这 $k$ 个样本中出现最多的类别标记作为预测结果；在回归任务中可使用“平均法”，即将这 $k$ 个样本的实值输出标记的平均值作为预测结果；还可基于距离远近进行加权平均或加权投票，距离越近的样本权重越大。

- 优点：精度高，对异常值不敏感，无数据输入假定
- 缺点：计算复杂度高，空间复杂度高
- 适用数据范围：数值型和标称型



### 多维缩放
在高维情形下会出现样本稀疏、计算距离困难等问题，即所谓的“维数灾难”。通常会通过降维来缓解维数灾难。
>维数简约：即通过某种数学变换将原始高维属性空间转变为一个低维“子空间”（subspace），在这个子空间中样本密度大幅提高，距离计算也变得更为容易。

若要求原始空间中样本之间的距离在低维空间中得以保持，即得到一种经典降维方法 --- 多维缩放(*Multiple Dimensional Scaling*)



D 为原始样本空间的距离矩阵，$D\in R^{m \times m}$

Z 为降维后的样本空间，$Z\in R^{d'\times m}$

B 为降维后样本的内积矩阵，即 $B = Z^TZ,b_{ij}=-\frac{1}{2}(dist^2_{ij}-dist^2_{i.}-dist^2_{.j}+dist^2_{..})$

其中：

$$
dist^2_{i.} = \frac{1}{m}\sum_{j=1}^{m}dist^2_{ij}\\
dist^2_{.j} = \frac{1}{m}\sum_{i=1}^{m}dist^2_{ij}\\
dist^2_{..} = \frac{1}{m}\sum_{i=1}^{m}\sum_{j=1}^{m}dist^2_{ij}
$$


### 主成分分析

>主成分分析（*Principal Component Analysis*，简称 PCA）是最常用的一种降维方法

- 最近重构性：样本点到这个超平面的距离都足够近
- 最大可分性：样本点在这个超平面上的投影能尽可能分开

**优化目标**
$$
min -tr(W^TXX^TW)\\
s.t. W^TW=I\\
$$
引入拉格朗日乘子可得：
$$
XX^TW=\lambda W
$$
于是，只需对协方差矩阵XXT进行特征值分解，将求得的特征值排序：$\lambda_1\geq \lambda_2\geq ... \lambda_d$，再取前 $d'$ 个特征值对应的特征向量构成 $W=（w_1,w_2,...,w_{d'}$，即为 主成分分析的解。



### 等度量映射

>流形学习： 是一类借鉴了拓扑流形概念的降维方法。“流形”是在局部与欧氏空间同胚的空间。

等度量映射（*Isometric Mapping*，简称 Isomap）是**流形学习**的一种。利用流形在局部上与欧氏空间同胚这个性质，对每个点基于欧氏距离找出其近邻点，然后就能建立一个近邻连接图，图中近邻点之间存在连接，而非近邻点之间不存在连接，于是，计算两点之间测地线距离的问题，就转变为计算近邻连接图上两点之间的最短路径问题（*Dijkstra 算法或 Floyd 算法*）。基于近邻距离逼近能获得低维流形上测地线距离很好的近似。



### 局部线性嵌入

>局部线性嵌入(*Locally Linear Embedding*，简称 LLE )试图保持邻域内样本之间的线性关系。假设样本 $x_i$的邻近点为{$x_j,x_k,x_l$} 则：

$$
x_i = w_{ij}x_j+w_{ik}x_k+w_{il}x_l
$$

求解推导过程 [参考](https://datawhalechina.github.io/pumpkin-book/#/chapter10/chapter10)

```python
# 导包
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
#解决中文显示问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 模拟数据

def setData():
    x = np.array([[1.0,1.1],[1.0,1.0],[0,0],[0,0.1]])
    y = ['A','A','B','B']
    
    return x,y
```

```python
x,y = setData()
```

```python
plt.scatter(x[:,0],x[:,1],c='r',marker='x')
```

```python
# 根据欧式距离公式实现的 knn
def classify(x,label,k,test):

    m = x.shape[0]
    # 扩展 test为 m 行
    ext_test = np.tile(test,(m,1))
    # 计算测试点到样本点的距离
    dist = np.sqrt(np.sum((ext_test-x)**2,axis=1))
    # 距离递增排序
    up_dist = np.argsort(dist)
    # 确定前 k 个点所在类别的出现频率
    classcount = {}
    for i in range(k):
        type_info = label[up_dist[i]]
        classcount[type_info] = classcount.get(type_info,0)+1
    up_type = sorted(classcount.items(),key=lambda x:x[1],reverse=True)
    
    return up_type[0][0]
```

```python
classify(x,y,3,[2,0.2])
```

### Code-MDS

```python
# 求解距离矩阵 D


def distMatr(x):
    cp_x = np.copy(x)
    m = cp_x.shape[0]
    dist = np.zeros((m, m))
    for i in range(m):
        ext = np.tile(cp_x[i, :], (m, 1))
        dist[i, :] = np.sqrt(np.sum((ext - cp_x)**2, axis=1))

    return dist


# MDS 降维
def MDSdemo(D, d):
    """
    D:样本的距离矩阵
    d:低维空间维数
    """
    cp_D = np.copy(D)
    m = cp_D.shape[0]

    dist2_i = np.tile((1 / m) * np.sum(cp_D**2, axis=1).reshape((m, 1)),
                      (1, m))
    dist2_j = np.tile((1 / m) * np.sum(cp_D**2, axis=0).reshape((1, m)),
                      (m, 1))
    dist2 = np.tile((1 / m**2) * np.sum(cp_D**2), (m, m))
    dist2_ij = cp_D**2

    # 计算矩阵 D 的内积矩阵 B
    B = (-1 / 2) * (dist2_ij - dist2_i - dist2_j + dist2)

    # 对矩阵 B 做特征值分解
    eigenvalue, featurevector = np.linalg.eig(B)

    # 选取 d 个最大特征值
    eigen = sorted(eigenvalue, reverse=True)[:d]
    arg = np.argsort(-eigenvalue)

    # d 个最大特征值构成对角矩阵
    A = np.eye(len(eigen)) * eigen

    # 特征向量矩阵
    V = featurevector[:, arg[:d]]

    Z = V @ np.sqrt(A)  # shape = (m, d)

    return Z
```

```python
Dist = distMatr(np.array([[-1,1,0,6],[-4,3,0,0.2],[1,0,2,-0.8],[2,3,-5,7]]))
```

```python
MDSdemo(Dist,2)
```

```python
np.sum(MDSdemo(D=Dist,d=2))
```

### Code-PCA

```python
# PCA 算法


## 对所有样本进行 中心化
def centralization(X):
    cp_x = np.copy(X)
    m = cp_x.shape[0]
    tmp = (1 / m) * np.sum(cp_x, axis=0)
    ext_tmp = np.tile(tmp, (m, 1))
    cp_x = cp_x - ext_tmp

    return cp_x


def PCAdemo(D, d):
    """
    D：样本集
    d：低维空间维数
    """
    # 1. 中心化样本
    cp_D = np.copy(D)
    X = centralization(cp_D)

    # 2. 计算样本的协方差矩阵
    covar_mat = X @ X.T

    # 3. 对协方差矩阵做特征值分解
    eigenvalue, featurevector = np.linalg.eig(covar_mat)

    # 选取 d 个最大特征值
    eigen = sorted(eigenvalue, reverse=True)[:d]
    arg = np.argsort(-eigenvalue)

    # 4. 取最大的 d 个特征值所对应的特征向量矩阵
    W = featurevector[:, arg[:d]]

    return W
```

### Code-LLE

```python
# 局部线性嵌入

def LLEdemo(D,k,d):
    """
    D:样本集
    k:近邻参数
    d:低维空间维数
    """
    x = np.copy(D)
    m = x.shape[0]
    W = np.zeros((m,m))
    I = np.ones((k-1,1))
    for i in range(m):
        # 确定 xi 的 k 邻近
        Q,dist = get_knn_d(x,k,x[i]) # Q中包含了 i
    
        # 求 W,若 j 不在 Q 中则 W_ij = 0
        Xi = (x[i] - x[Q[1:]]).T
        pinv_x = np.linalg.pinv(Xi.T@Xi)
        W[i,Q[1:]] = ((pinv_x@I)/(I.T@pinv_x@I)).T
        
    # 求 M
    I = np.ones((m,1))
    M = (I - W).T@(I - W)
    
    # 对 M 做特征值分解
    eigenvalue, featurevector = np.linalg.eig(M)
    
    # 选取 最小 d 个特征值
    eigen = sorted(eigenvalue)[:d]
    arg = np.argsort(eigenvalue)
    Z = featurevector[:, arg[:d]]
    
    return Z
```
